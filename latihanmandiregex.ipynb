```python
import re
from datetime import datetime

text = """Pada tanggal 1945-08-17 Indonesia merdeka. Indonesia memiliki beberapa pahlawan
nasional, seperti Pangeran Diponegoro (TL: 1785-11-11), Pattimura (TL: 1783-06-08) dan Ki
Hajar Dewantara (1889-05-02)."""

tgl = re.findall(r'\d{4}-\d{2}-\d{2}', text)

# Tanggal saat ini
inihari = datetime.now()
for tgl_str in tgl:
    tgl_obj = datetime.strptime(tgl_str, '%Y-%m-%d')
    jumlah = inihari - tgl_obj
    print(f"{tgl_obj.strftime('%d-%m-%Y')} selisih {jumlah.days} hari")
```

```python
import re
import random
import string

def buat_kata_sandi(panjang=8):
  sumber_karakter = string.ascii_letters + string.digits
  kata_sandi_acak = ''.join((random.choice(sumber_karakter) for _ in range(panjang)))
  return kata_sandi_acak

def utama():
  nama_file_email = input("Masukkan nama file yang berisi alamat email: ")
  try:
    with open(nama_file_email, "r") as file_terbuka:
      for baris in file_terbuka:
        nama_pengguna = re.split("@", baris)[0].strip()
        cocok_email = re.findall(r"\S+@\S+", baris)
        if cocok_email:
          alamat_email = cocok_email[0]
          kata_sandi_acak = buat_kata_sandi()
          print(f"Email: {alamat_email}, Nama Pengguna: {nama_pengguna}, Kata Sandi: {kata_sandi_acak}")
  except FileNotFoundError:
    print(f"File '{nama_file_email}' tidak ditemukan. Silakan periksa kembali nama file Anda.")

if __name__ == "__main__":
  utama()
```
